```python
import os
import requests
import json
from pprint import pprint
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("SERP_API_KEY")
if not api_key:
    raise ValueError('SERP_API_KEY is not set in the environment varibles.')
```

```python
query = "Lithium Battery"
base_url = f"https://serpapi.com/search?engine=google_patents&q={query}&api_key={api_key}"
```

```python
response = requests.get(base_url)
if response.status_code == 200:
    data = response.json()
    
    patents = data.get("organic_results", [])
    with open("files/filespatents.json", "w") as file:
        json.dump(patents, file, indent=4)
```

```python
patent_url = f"https://serpapi.com/search.json?engine=google_patents_details&patent_id=patent%2FCN109659487B%2Fen&api_key={api_key}"
patents_response = requests.get(patent_url)
if patents_response.status_code == 200:
    patent_data = patents_response.json()

    with open("files/patent_details.json", "w") as file:
        json.dump(patent_data, file, indent=4)
```

```python
pprint(patent_data.get("abstract", ""))
```

```python
dir_files = os.listdir("results")

for file in dir_files:
    if file.endswith(".json"):
        with open(os.path.join("results", file), "r") as f:
            content = json.load(f)
            pprint(content.get("abstract", ""))
            print("\n" + "=" * 40 + "\n")
```
